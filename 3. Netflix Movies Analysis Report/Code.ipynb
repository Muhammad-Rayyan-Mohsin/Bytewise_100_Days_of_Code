```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
df = pd.read_csv('netflix_data.csv')

netflix_subset = df[df['type'] == 'TV Show']

netflix_movies = df[df['type'] == 'Movie']

netflix_movies = netflix_movies[['title', 'country', 'genre', 'release_year', 'duration']]

short_movies = netflix_movies[netflix_movies['duration'] < 60]

print(short_movies)

print("\nFrom the output we can deduce that most short movies are usually from United States and from the genre of Documentaries.")
```

__________________________________________________________________________________________________________________________________________________________________

```python
print("\nTo further validate this information, we can use the describe function on the relevant columns.")
print("\nGenre Description:")
print(short_movies['genre'].describe())
print("\n\nCountry Description:")
print(short_movies['country'].describe())
print()
print("The above provided evidence could possibly give us the potential contributing factors (Genre and Country) to the short duration of the movies.")
```

__________________________________________________________________________________________________________________________________________________________________

```python
colors = []
for index, row in netflix_movies.iterrows():
    if row['genre'] == 'Children':
        colors.append('red')
    elif row['genre'] == 'Documentaries':
        colors.append('blue')
    elif row['genre'] == 'Stand-Up':
        colors.append('green')
    else:
        colors.append('yellow')

fig = plt.figure()
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title('Movie Duration by Year of Release')

red_patch = mpatches.Patch(color='red', label='Children')
blue_patch = mpatches.Patch(color='blue', label='Documentaries')
green_patch = mpatches.Patch(color='green', label='Stand-Up')
yellow_patch = mpatches.Patch(color='yellow', label='Other')

plt.legend(handles=[red_patch, blue_patch, green_patch, yellow_patch], title="Genres")

plt.show()
```
